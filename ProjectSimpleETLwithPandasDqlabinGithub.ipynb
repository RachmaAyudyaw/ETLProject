Transform
Transform merupakan proses melakukan transformasi data, atau perubahan terhadap data.
Umumnya seperti:
1. Merubah nilai dari suatu kolom ke nilai baru,
2. Menciptakan kolom baru dengan memanfaatkan kolom lain,
3. Transpose baris menjadi kolom (Atau sebaliknya),
4. Mengubah format data ke bentuk yang lebih standart (contohnya, kolom date. maupun datetime yang biasanya memiliki nilai yang tidak standart maupun nomor HP yang biasanya memiliki nilai yang tidak sesuai format standardnya) dan lainnya

**Transform bagian 2 - kota**
buat kolom baru bernama city yang didapat dari kolom address. diasumsikan bahwa kota merupakan sekumpulan karakter yang terdapat setelah no jalan diikuti dengan \n dan sebelum karakter koma ,

```python
df_participant['city']=df_participant['address'].str.extract(r'(?<=\n)(\w.+)(?=,)')
df_participant[['address','city']]
```

(?<=\n)(\w.+)(?=,) mempunyai arti:
* '?': Mencocokkan karakter 0 atau 1 setelahnya
*   \w: Mencocokkan karakter angka [0-9], alfabet [a-zA-Z] dan underscore (_)
*   '.': Mencocokkan sembarang karakter (kecuali enter/newline) tanpa pengulangan
*   '+': pengulangan 1 atau lebih karakter


* (?<=\n) : temukan karakter \n lalu abaikan semua karakter sebelum tanda / simbol \n
* (\w.+) : berarti ambil karakter non simbol dan pencarian akan dihentikan saat bertemu simbol koma pada kombinasi simbol (?=,)

**Transform Bagian V - Nama Tim**
Buatlah kolom baru bernama team-name. Nama tim merupakan gabungan nilai dari kolom first_name, last_name, country dan institute dengan ketentuan:
1. Ambil awalan first_name yang digabung dengan last_name (singkatan nama): Contoh: Danang Qornain menjadi DQ
2. Tambahkan nilai dari kolom negara (country)
3. Tambahkan awalan setiap kata dari kolom institue yang digabungkan. Contoh: Universitas Digitalent Kominfo menjadi UDK



```python
#Buat sebuah fungsi
def TeamName(dataframe):
  #1. singkatan dari nama depan dan nama belakang dengan mengambil huruf pertama (indeks 0)
  abbrev_name="%s%s"%(dataframe['first_name'][0],dataframe['last_name'][0])
  #2. ambil nilai dari kolom negara/country
  country=dataframe['country']
  #3. singkatan dari value di kolom institute
  abbrev_institute='%s'%(''.join(list(map(lambda word: word[0], dataframe['institute'].split()))))
  return "%s-%s-%s"%(abbrev_name,country,abbrev_institute)

#aplikasikan fungsi yang telah dibuat ke dalam dataframe
df_participant['team_name']=df_participant.apply(TeamName,axis=1)

#tampilkan hasil
df_participant[['first_name','last_name','country','institute','team_name']]
```

Keterangan:
1. pada sintaks abbrev_name = "%s%s"%(dataframe['first_name'][0], dataframe['last_name'][0]) berarti mengambil karakter index ke 0 dari firstname dan index ke 0 dari lastname
2. sintaks country = dataframe['country'] mengambil nilai pada kolom country dan dimasukkan ke variabel country
3. sintaks abbrev_institute = '%s'%(''.join(list(map(lambda word: word[0),dataframe['institute'].split()))) , yaitu pertama melakukan split pada dataframe['institute'] sehingga misalkan ada Universitas Digitalent Kominfo akan di split menjadi Universitas-Digitalent-Kominfo, kemudian fungsi lambda word: word[0] adalah mengambil index pertama dari tiap kata yang sudah di split di dataframe['Insitute'] tadi sehinga menjadi: U D K, lalu fungsi join untuk menggabungkanya menjadi 1 yaitu UDK



**Transform Bagian 6- Email**
format email : xxyy@aa.bb[ac/com].[cc]
keterangan:
* xx -> nama depan (first_name) dalam lowercase
* yy -> nama belakang (last_name) dalam lowercase
* aa -> nama institusi

untuk nilai bb dan cc mengikuti dari aa. Aturannya:
* Jika institusi nya merupakan Universitas, maka bb->gabungan dari huruf pertama pada setiap kata dari nama Universitas dalam lowercase kemudian diikuti ac yang menandakan akademi/institusi belajar dan diikuti dengan pattern cc

* Jika institusi bukan merupakan Universitas, maka bb -> gabungan dari huruf pertama pada setiap kata dari nama Universitas dalam lowercase kemudian diikuti dengan .com. perlu diketahui bahwa pattern cc tidak berlaku pada pattern ini

untuk nilai cc:
* jika banyaknya kata pada negara tersebut lebih dari 1 maka ambil singkatan dari negara tersebut dalam lowercase
* Namun, jika banyaknya kata hanya 1 maka ambil 3 huruf terdepan dari negara tersebut dalam lowercase













**Contoh 1**
* Nama depan: Citra
* Nama belakang: Nurdiyanti
* Institusi: UD Prakasa Mandasari
* Negara: Georgia
maka emailnya citranurdiyanti@upm.com

**Contoh 2**
* Nama depan: Aris
* Nama belakang: Setiawan
* Institusi: Universitas Diponegoro
* Negara: Korea Utara
maka, emailnya arissetiawan@ud.ac.ku

```python
def email(col):
  #Ambil nama depan dan nama belakang menjadi lowercase
  first_name_lower=col['first_name'].lower()
  last_name_lower=col['last_name'].lower()

  #ambil masing-masing singkatan institute
  institute=''.join(list(map(lambda x: x[0],col['institute'].lower().split())))

  #cek apakah kolom institute mengandung kata universitas?
  if 'Universitas' in col['institute']:
    #jika ya lalu cek apakah kolom country mengandung lebih dari satu kata?
    if len(col['country'].split())>1:
      #jika ya maka ambil singkatan(karakter depan) dari masing-masing kolom county lalu di ubah ke lowercase
      country=''.join(list(map(lambda x:x[0],col['country'].lower().split())))
    else:
      #jika tidak ambil karakter awal lalu ubah ke lowercase
      country=col['country'][:3].lower()
    #return nilai pengkodisian universitas
    return "%s%s@%s.ac.%s"%(first_name_lower,last_name_lower,institute,country)
  #return fungsi email
  return "%s%s@%s.com"%(first_name_lower,last_name_lower,institute)

#aplikasikan fungsi email dan buat kolom baru bernama email
df_participant['email']=df_participant.apply(email,axis=1)

#tampilkan hasil
df_participant[['first_name','last_name','institute','country','email']]
```

**Transform Bagian 8 - Tanggal daftar kompetisi**
> selain punya aturan mengenai format DATE, MySQL juga memberi aturan pada data yang bertipe DATETIME yaitu YYYY-MM-DD HH:mm:ss dengan keterangan:
* YYYY: 4 digit yang menandakan tahun
* MM: 2 digit yang menandakan bulan
* DD: 2 digit yang menandakan tanggal
* HH: 2 digit yang menandakan jam
* mm: 2 digit yang menandakan menit
* ss: 2 digit yang menandakan detik
contohnya: 2021-04-07 15:10:55

karena format data pada register at belum sesuai format yang seharusnya. maka dari itu ubahlah register_time ke format DATETIME sesuai dengan aturan dari MySQL







```python
df_participant['register_at']=pd.to_datetime(df_participant['register_time'],unit='s')
df_participant[['register_time','register_at']]
```
