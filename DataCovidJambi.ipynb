```python
import requests
resp=requests.get('https://data.covid19.go.id/public/api/prov_detail_JAMBI.json')
#respon api 200 menyatakan bahwa permintaan anda berhasil dipenuhi dan konten yang diminta tersedia untuk diekstrak
#dari atribut header dapat diketahui bahwa konten yang diminta dalam bentuk application/json, yaitu berkas json
print(resp)
print(resp.headers)
```

```python
#mengekstrak konten dengan menggunakan method json()
cov_jambi_row=resp.json()
print(cov_jambi_row)
```

```python
print('Length of cov_jambi_row: {}'.format(len(cov_jambi_row)))
print('Komponen cov_jambi:%s.'%cov_jambi_row.keys())
```

Memperoleh informasi yang lebih lengkap
Informasi umum mengenai COVID-19 di Jambi telah didapatkan. Namun informasi akan lebih lengkap jika kita memiliki data perkembangan COVID-19 dari waktu ke waktu.
Data historis perkembangan COVID-19 tersebut tersimpan dengan nama list_perkembangan. Ubah data cov_jambi_row dari bertipe dict menjadi pandas dataframe. Kemudian simpan hasilnya sebagai obyek bernama cov_jambi. amati struktur cov_jambi menggunakan method info() dan head()

```python
#ubah cov_jambi_row dari dict menjadi pandas dataframe
import pandas as pd
cov_jambi=pd.DataFrame(cov_jambi_row['list_perkembangan'])
print('Info cov_jambi:',cov_jambi.info())
print('\nLima data teratas cov_jambi:\n')
cov_jambi.head(5)
```

Ada beberapa tahapan yang akan anda lakukan untuk memperbaiki data cov_jambi yaitu:
1. Menghapus kolom DIRAWAT_OR_ISOLASI dan semua kolom yang berisi nilai kumulatif
2. Merubah format penulisan semua kolom menjadi huruf kecil
3. Mengganti nama kolom kasus menjadi kasus_baru
4. Memperbaiki tipe data pada kolom tanggal menggunakan pd.to_datetime
Simpan hasil perubahan dengan nama cov_jambi_tidy

```python
data=[]
for item in cov_jambi.columns:
  if item.startswith('AKUMULASI') | item.startswith('DIRAWAT'):
    data.append(item)

cov_jambi_tidy=cov_jambi.drop(columns=data)
```

```python
#print(cov_jambi_tidy)
cov_jambi_tidy=cov_jambi_tidy.rename(columns=str.lower).rename(columns={'kasus':'kasus_baru'})
cov_jambi_tidy
```

```python
cov_jambi_tidy['tanggal']=pd.to_datetime(cov_jambi_tidy['tanggal']*1e6,unit='ns')
cov_jambi_tidy['tanggal']
```

```python
tanggal=cov_jambi_tidy['tanggal'].dt.strftime('%d-%m-%Y')
print(tanggal,cov_jambi_tidy['tanggal'])
```

Visualisasi

```python
import matplotlib.pyplot as plt
plt.clf()
fig,ax=plt.subplots(figsize=(10,5))
ax.bar(data=cov_jambi_tidy,x='tanggal',height='kasus_baru')
plt.show()
```

```python
import matplotlib.dates as mdates
plt.clf()
fig,ax=plt.subplots(figsize=(10,5))
ax.bar(data=cov_jambi_tidy, x='tanggal', height='kasus_baru', color='salmon')
fig.suptitle('Kasus harian Positif COVID-19 di Jambi',y=1.00,fontsize=16,fontweight='bold',ha='center')
ax.set_title('Terjadi pelonjakan kasus di bulan Juli 2021 diduga karena varian delta plus',fontsize=10)
ax.set_xlabel('')
ax.set_ylabel('Jumlah kasus')
ax.text(1,-0.3,'sumber: covid19.go.id',color='blue',ha='right',transform=ax.transAxes)
ax.set_xticklabels(ax.get_xticks(),rotation=90)

ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
```
